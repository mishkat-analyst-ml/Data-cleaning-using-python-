# check Null Values 

```python
df.isnull().sum()
```

### postal code have 11 null values so we will fill it by mean,median OR 0 

```python
df['Postal Code'].fillna(df['Postal Code'].median(), inplace=True)
```

### the date coulmns is in str so we convert it into DATE FORMAT

```python
df['Order Date']=pd.to_datetime(df['Order Date'],format="%d/%m/%Y")
df['Ship Date']=pd.to_datetime(df['Ship Date'],format="%d/%m/%Y")
```

### removing strips from all coulmns

```python
df=df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

### check duplicates 

```python
df.duplicated().sum()
```

```python
df=df.drop(['Product ID','Segment'],axis=1)
```

```python
df['Customer ID'].unique()
```

### save clean DATASET

```python
df.to_csv("cleaned_data.csv",index=False)
```
